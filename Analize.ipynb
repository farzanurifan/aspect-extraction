```python
import pandas as pd

def analize(label, tipe, positif):
    df = pd.read_csv('Results/Categorization/'+ tipe +'.csv')
    
    tp = 0
    tn = 0
    fp = 0
    fn = 0
    kosong = 0
    
    for index in range(0, len(df)):
        labels = df['label'][index].split(',')
        
        if type(df['predict'][index]) != float:
            preds = df['predict'][index].split('|')

            if label in labels and label in preds:
                tp += 1
            elif label not in labels and label not in preds:
                tn += 1
            elif label not in labels and label in preds:
                if positif:
                    print(df['review'][index], '~')
                    print()
                    print('label\t:', df['label'][index])
                    print('target\t:', df['target'][index])
                    print('term\t:', df['terms'][index])
#                     error = []
#                     for i, aspek in enumerate(df['pred_multiple'][index].split('|')):
#                         if aspek == label:
#                             error.append(df['terms'][index].split('|')[i])
#                     print(' | '.join(error))
                    print('pred\t:', df['predict'][index])
                    print('_____________________________________________________________________________________________________')
                    print()
                fp += 1
            elif label in labels and label not in preds:
                if not positif:
                    print(df['review'][index], '~')
                    print()
                    print('label\t:', df['label'][index])
                    print('target\t:', df['target'][index])
                    print('term\t:', df['terms'][index])
#                     error = []
#                     for i, aspek in enumerate(df['pred_multiple'][index].split('|')):
#                         if aspek != label:
#                             error.append(df['terms'][index].split('|')[i])
#                     print(' | '.join(error))
                    print('pred\t:', df['predict'][index])
                    print('_____________________________________________________________________________________________________')
                    print()
                fn += 1
        elif label in labels:
            fn += 1
            if not positif:
                print(df['review'][index], '~')
                print()
                print('label\t:', df['label'][index])
                print('target\t:', df['target'][index])
                print('---')
                print('')
        else:
            tn += 1
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1 = 2 * ((precision * recall) / (precision + recall))
    print(tp, tn, fp, fn, precision, recall, f1)
    return tp, tn, fp, fn, precision, recall, f1
```
