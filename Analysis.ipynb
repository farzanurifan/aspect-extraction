```python
import pandas as pd

dp = pd.read_csv("hasil_dp_2.csv")
dp_c = pd.read_csv("hasil_dp_2_clause.csv")
```

```python
df = pd.DataFrame(columns=['review','target','dp','dp_clause'])

for x in range(0, len(dp)):
    review = dp['0'][x]
    target = dp['aspect'][x]
    first = dp['prediction'][x]
    second = dp_c['prediction'][x]
    if first != second and (isinstance(first, str) or isinstance(second, str)):
        df = df.append({'review': review, 'target': target, 'dp': first, 'dp_clause': second}, ignore_index=True)
        print(target, first, second)

df.to_csv('difference_2.csv')
```
