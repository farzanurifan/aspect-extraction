```python
import pandas as pd
import math

def get_array(tipe):
    df = pd.read_csv('input_'+ tipe +'.csv')
    arr = []
    for i in range(0, 10):
        start = int(len(df)*i/10)
        end = int(len(df)*(i+1)/10)
        arr.append(df['predict'][start:end])
    return arr

def hitung_nilai(arr):
    res = 0
    for a in arr:
        if a == 'positive':
            res += 1
        elif a == 'negative':
            res -= 1
        else:
            print('err')
    return res

def moora(bobot_1, bobot_2, bobot_3, bobot_4):
    ambience = get_array('AMBIENCE')
    food = get_array('FOOD')
    service = get_array('SERVICE')
    prices = get_array('PRICES')

    restaurant = []
    for i in range(0, 10):
        data = [hitung_nilai(ambience[i]),
                hitung_nilai(food[i]),
                hitung_nilai(service[i]),
                hitung_nilai(prices[i])]
        
        restaurant.append(data)

    sum1, sum2, sum3, sum4 = 0, 0, 0, 0
    for res in restaurant:
        kriteria_1 = res[0]
        kriteria_2 = res[1]
        kriteria_3 = res[2]
        kriteria_4 = res[3]
        sum1 += kriteria_1 * kriteria_1
        sum2 += kriteria_2 * kriteria_2
        sum3 += kriteria_3 * kriteria_3
        sum4 += kriteria_4 * kriteria_4

    sqrt1 = math.sqrt(sum1)
    sqrt2 = math.sqrt(sum2)
    sqrt3 = math.sqrt(sum3)
    sqrt4 = math.sqrt(sum4)

    normalized = []
    for res in restaurant:
        normalized_1 = res[0] / sqrt1
        normalized_2 = res[1] / sqrt2
        normalized_3 = res[2] / sqrt3
        normalized_4 = res[3] / sqrt4

        normalized.append([normalized_1,
                          normalized_2,
                          normalized_3,
                          normalized_4])
    weighted = []
    for normal in normalized:
        weighted1 = normal[0] * bobot_1
        weighted2 = normal[1] * bobot_2
        weighted3 = normal[2] * bobot_3
        weighted4 = normal[3] * bobot_4

        weighted.append([weighted1,
                      weighted2,
                      weighted3,
                      weighted4])

    summed = []
    for index, weight in enumerate(weighted):
        sum_all = weight[0] + weight[1] + weight[2] + weight[3]
        summed.append((index + 1, sum_all))


    def getKey(item):
        return item[1]

    sorted_data = sorted(summed, key=getKey, reverse=True)
    
    return sorted_data

moora(0.5,0.1,0.1,0.3)
```
