```python
import pandas as pd
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

custom_sent_tokenizer = PunktSentenceTokenizer()

def chunk(sentence):
    try:
        tokenized = custom_sent_tokenizer.tokenize(sentence)
        for i in tokenized:
            hasil = []
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
#             chunkGram = r"""Chunk: {<DT>?<NN.*|JJ>?<IN>?<PRP.*>?<NN.*>}"""
            chunkGram = r"""Chunk: {<NN.*>?<IN>?<PRP.*>?<NN.*>}"""
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            for subtree in chunked.subtrees(filter=lambda t: t.label() == 'Chunk'):
                panjang = len(subtree)
                rangkaian_kata = ''
                for i in range(0,panjang):
                    if i == 0:
                        rangkaian_kata += subtree[i][0]
                    else:
                        rangkaian_kata += ' ' + subtree[i][0]
                hasil.append(rangkaian_kata)
            return hasil

    except Exception as e:
        print(str(e))

def combine_chunk(word, chunked):
    for chunked_word in chunked:
        if word in chunked_word:
            return chunked_word
    return word
```

```python
from pycorenlp import StanfordCoreNLP
import json

nlp = StanfordCoreNLP('http://localhost:9000')
dependency_parser = nlp.annotate

document = open("Nokia 6610.txt", "r")

# Dependency relations describing relations between opinion words and targets
# tak tambahi amod & nsubj
MR = ['mod', 'pnmod', 'subj', 's', 'obj', 'obj2','desc', 'amod', 'nsubj']
CONJ = ['conj', 'compound']

same_mod = ['mod', 'pnmod', 'amod']
same_bj = ['subj', 'obj', 'nsubj']

# Initial Opinion Words
ow_set = set(line.strip() for line in open('initial_opinion.txt'))

aspects = []
a_set = set(aspects)
a_set_temp = set(aspects)

while(True):
    flag_ow = len(ow_set) 
    flag_a = len(a_set)
    # Loop 1
    for doc in document:
        if len(doc.split("##")) == 1:
            continue
        sentence = doc.split("##")[1]
        for all_aspect in doc.split("##")[0].split(", "):
            aspek = all_aspect.split("[")[0]
            print(all_aspect)
            print(aspek)
            
        # Parse
        result = dependency_parser(sentence, properties={"outputFormat": "json","annotators": "pos, depparse"})
        hasil_chunk = chunk(sentence)

        for res in result['sentences']:

            # Rule 1.1
            for data in res['basicDependencies']:
                w1 = data['governorGloss']
                w2 = data['dependentGloss']
                w1_pos = res['tokens'][data['governor']-1]['pos']
                w2_pos = res['tokens'][data['dependent']-1]['pos']
                dep = data['dep']

                target = ''
                if w1 in ow_set:
                    if w2_pos == 'NN':
                        target = w2
                elif w2 in ow_set:
                    if w1_pos == 'NN':
                        target = w1
                if target != '':
                    if dep in MR:
                        a_set_temp.add(target)
                        a_set.add(combine_chunk(target, hasil_chunk))

            # Rule 1.2
            for data in res['basicDependencies']:
                w1 = data['governorGloss']
                w2 = data['dependentGloss']
                w1_pos = res['tokens'][data['governor']-1]['pos']
                w2_pos = res['tokens'][data['dependent']-1]['pos']
                dep = data['dep']

                H = ''
                if w1 in ow_set:
                    H = w2
                elif w2 in ow_set:
                    H = w1

                if H != '':
                    if dep in MR:
                        for data2 in res['basicDependencies']:
                            w1 = data2['governorGloss']
                            w2 = data2['dependentGloss']
                            w1_pos = res['tokens'][data2['governor']-1]['pos']
                            w2_pos = res['tokens'][data2['dependent']-1]['pos']
                            dep = data2['dep']

                            target = ''
                            if w1 == H:
                                if w2_pos == 'NN':
                                    target = w2
                            elif w2 == H:
                                if w1_pos == 'NN':
                                    target = w1
                            if target != '':
                                if dep in MR:
                                    a_set_temp.add(target)
                                    a_set.add(combine_chunk(target, hasil_chunk))


            # Rule 4.1
            for data in res['basicDependencies']:
                w1 = data['governorGloss']
                w2 = data['dependentGloss']
                w1_pos = res['tokens'][data['governor']-1]['pos']
                w2_pos = res['tokens'][data['dependent']-1]['pos']
                dep = data['dep']

                target = ''
                if w1 in ow_set:
                    if w2_pos == 'JJ':
                        target = w2
                elif w2 in ow_set:
                    if w1_pos == 'JJ':
                        target = w1
                if target != '':
                    if dep in CONJ:
                        ow_set.add(target)

            # Rule 4.2
            for data in res['basicDependencies']:
                w1 = data['governorGloss']
                w2 = data['dependentGloss']
                w1_pos = res['tokens'][data['governor']-1]['pos']
                w2_pos = res['tokens'][data['dependent']-1]['pos']
                dep1 = data['dep']

                H = ''
                if w1 in ow_set:
                    H = w2
                elif w1 in ow_set:
                    H = w2

                if H != '':
                    for data2 in res['basicDependencies']:
                        w1 = data2['governorGloss']
                        w2 = data2['dependentGloss']
                        w1_pos = res['tokens'][data2['governor']-1]['pos']
                        w2_pos = res['tokens'][data2['dependent']-1]['pos']
                        dep2 = data2['dep']

                        target = ''
                        if w1 == H:
                            if w2_pos == 'JJ':
                                target = w2
                        elif w2 == H:
                            if w1_pos == 'JJ':
                                target = w1
                        if target != '':
                            if dep1 in same_bj:
                                if dep2 in same_bj:
                                    ow_set.add(target)
                            elif dep1 in same_mod:
                                if dep2 in same_mod:
                                    ow_set.add(target)


    # Loop 2
    for doc in document:        
        if len(doc.split("##")) == 1:
            continue
        sentence = doc.split("##")[1]
        # Parse
        result = dependency_parser(sentence, properties={"outputFormat": "json","annotators": "pos, depparse"})
        hasil_chunk = chunk(sentence)

        for res in result['sentences']:

            # Rule 3.1
            for data in res['basicDependencies']:
                w1 = data['governorGloss']
                w2 = data['dependentGloss']
                w1_pos = res['tokens'][data['governor']-1]['pos']
                w2_pos = res['tokens'][data['dependent']-1]['pos']
                dep = data['dep']

                target = ''
                if w1 in a_set_temp:
                    if w2_pos == 'NN':
                        target = w2
                elif w2 in a_set_temp:
                    if w1_pos == 'NN':
                        target = w1
                if target != '':
                    if dep in CONJ:
                        a_set_temp.add(target)
                        a_set.add(combine_chunk(target, hasil_chunk))

            # Rule 3.2
            for data in res['basicDependencies']:
                w1 = data['governorGloss']
                w2 = data['dependentGloss']
                w1_pos = res['tokens'][data['governor']-1]['pos']
                w2_pos = res['tokens'][data['dependent']-1]['pos']
                dep1 = data['dep']

                H = ''
                if w1 in a_set_temp:
                    H = w2
                elif w2 in a_set_temp:
                    H = w1

                if H != '':
                    for data2 in res['basicDependencies']:
                        w1 = data2['governorGloss']
                        w2 = data2['dependentGloss']
                        w1_pos = res['tokens'][data2['governor']-1]['pos']
                        w2_pos = res['tokens'][data2['dependent']-1]['pos']
                        dep2 = data2['dep']

                        target = ''
                        if w1 == H:
                            if w2_pos == 'NN':
                                target = w2
                        elif w2 == H:
                            if w1_pos == 'NN':
                                target = w1
                        if target != '':
                            if dep1 in same_bj:
                                if dep2 in same_bj:
                                    a_set_temp.add(target)
                                    a_set.add(combine_chunk(target, hasil_chunk))
                            elif dep1 in same_mod:
                                if dep2 in same_mod:
                                    a_set_temp.add(target)
                                    a_set.add(combine_chunk(target, hasil_chunk))

            # Rule 2.1
            for data in res['basicDependencies']:
                w1 = data['governorGloss']
                w2 = data['dependentGloss']
                w1_pos = res['tokens'][data['governor']-1]['pos']
                w2_pos = res['tokens'][data['dependent']-1]['pos']
                dep = data['dep']

                target = ''
                if w1 in a_set_temp:
                    if w2_pos == 'JJ':
                        target = w2
                elif w2 in a_set_temp:
                    if w1_pos == 'JJ':
                        target = w1
                if target != '':
                    if dep in MR:
                        ow_set.add(target)

            # Rule 1.2
            for data in res['basicDependencies']:
                w1 = data['governorGloss']
                w2 = data['dependentGloss']
                w1_pos = res['tokens'][data['governor']-1]['pos']
                w2_pos = res['tokens'][data['dependent']-1]['pos']
                dep = data['dep']

                H = ''
                if w1 in a_set_temp:
                    H = w2
                elif w2 in a_set_temp:
                    H = w1

                if H != '':
                    if dep in MR:
                        for data2 in res['basicDependencies']:
                            w1 = data2['governorGloss']
                            w2 = data2['dependentGloss']
                            w1_pos = res['tokens'][data2['governor']-1]['pos']
                            w2_pos = res['tokens'][data2['dependent']-1]['pos']
                            dep = data2['dep']

                            target = ''
                            if w1 == H:
                                if w2_pos == 'JJ':
                                    target = w2
                            elif w2 == H:
                                if w1_pos == 'JJ':
                                    target = w1
                            if target != '':
                                if dep in MR:
                                    ow_set.add(target)
    print("-")
    if flag_ow == len(ow_set):
        if flag_a == len(a_set):
            break
    else:
        document = open("Nokia 6610.txt", "r")
print(a_set_temp)
print(a_set)
```
